```python
import pandas as pd

# Dados iniciais

data = {'NCM': [1001, 1002, 1003,10011, 100112, 100113, 100114, 10012, 100121, 100122, 100122],
        'DESCRIÇÃO': ['Produto A', 'Produto B', 'Produto C', 'produto agricolas', 'milho', 'feijão','arroz', 'produtos industrializados', 'macarrao', 'massa de milho', 'massa de trigo',],
        'TAXA': [10.5, 15.2, 8.7, 0.0, 0.8, 0.7, 1.8, 0.0, 0.8, 0.5, 0.2]}

df = pd.DataFrame(data)

print(df)

# Função para gerar a coluna "Complemento"
df = pd.DataFrame(data)

# Função para gerar a coluna "Complemento"
def gerar_complemento(df):
    # Dicionário para armazenar a descrição dos códigos de 5 dígitos
    complemento_dict = {}

    # Itera sobre cada linha do DataFrame
    for index, row in df.iterrows():
        codigo_str = str(row['NCM'])
        # Pega os primeiros 5 dígitos
        if len(codigo_str) == 5:
            complemento_dict[codigo_str] = row['DESCRIÇÃO']

    # Cria a coluna "Complemento"
    def concatena_complemento(row):
        codigo_str = str(row['NCM'])
        if len(codigo_str) == 5:
            # Para códigos de 5 dígitos, apenas a descrição
            return complemento_dict.get(codigo_str, row['DESCRIÇÃO'])
        else:
            # Para outros códigos, concatena com a descrição do código de 5 dígitos
            prefixo = codigo_str[:5]
            if prefixo in complemento_dict:
                return complemento_dict[prefixo] + ' ' + row['DESCRIÇÃO']
            return row['DESCRIÇÃO']

    df['DESC_COMPL'] = df.apply(concatena_complemento, axis=1)

    return df

# Chama a função
df_resultado = gerar_complemento(df)

# Exibe o resultado
print(df_resultado)
```
